Download Tom Sawyer dataset
Compute Tokens, Types and TTR
Check if Zipf's law hold true for meanings and lengths
plot heaps' law

//The Dataset has been downloaded as TomSawyer.txt

```python
#imports
import nltk
import operator
nltk.download('punkt')
import matplotlib.pyplot as plt

from nltk.tokenize import word_tokenize


#opening the file
filePointer = open('TomSawyer.txt','r')
#print (filePointer.read())

dataSet=[]

#reading the file and word tokenizing it to a list

dataSet=word_tokenize(filePointer.read())
#print(dataSet)

#closing the file to dataset
filePointer.close()

print("Total Number of tokens in the dataset is " + str(len(dataSet)))

#calculating the number  of types
typesInDataSetAsSet=set(dataSet)
typesOfTheDataSet=list(typesInDataSetAsSet)

print("Total Number of types in the dataset is "+ str(len(typesOfTheDataSet)))

TTR = (len(typesOfTheDataSet))/(len(dataSet))

print("TTR is " +str(TTR))


```

```python
#trying to find the frequency distribution

diction = {}
for x in typesOfTheDataSet:
    diction[x]=0;
    
for x in dataSet:
    diction[x] = diction[x] + 1
    
#print(diction)

sortedDiction= sorted(diction.items(),key=operator.itemgetter(1), reverse=True)

#print(sortedDiction)

sortedDictionary=dict(sortedDiction)

#print(sortedDictionary)

#print(sortedDictionary[','])

frequencyXRank = []
rankList=[]
frequencyList=[]

rank=1

for key in sortedDictionary.keys():
    frequencyXRank.append((rank*sortedDictionary[key]))
    frequencyList.append(sortedDictionary[key])
    rankList.append(rank)
    rank=rank+1
    
#print(frequencyList)
#print(len(rankList))

print('Rank on x axis (semiLog), Frequency  on Y axis')

#plt.plot(rankList,frequencyList )

fig, ax = plt.subplots()
ax.semilogx(rankList,frequencyList)


plt.show()
```

```python
frequencyList=[]
lengthList=[]

for key in sortedDictionary.keys():
    frequencyList.append(sortedDictionary[key])
    lengthList.append(len(key))
    
    
#print(frequencyList)
#print(len(rankList))

print('Length on x axis, Frequency  on Y axis')

#plt.plot(rankList,frequencyList )


plt.plot(lengthList,frequencyList)


plt.show()
```

```python
print("Heaps Law")
```

```python
V=[]
N=[]

tempList=[]

counter = 1

for x in dataSet:
    tempList.append(x)
    counter= counter +1
    if(counter%200==0):
        N.append(len(tempList))
        V.append(len(set(tempList)))
    
print("Number of tokens on X axis and Size of Vocabulary on Y axis")
plt.title('Heaps Law')

plt.plot(N,V)
plt.show()
```
